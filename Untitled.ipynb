```python



# gca stands for 'get current axis'
ax = plt.gca()

df_fr.plot(kind='line',x='date',y='France',ax=ax)
df_us.plot(kind='line',x='date',y='US', color='red', ax=ax)
df_in.plot(kind='line',x='date',y='India', color='orange', ax=ax)

plt.show()
```

```python
df = pd.read_sql("SELECT continent, total_deaths_per_million FROM covid WHERE date = '2020-04-07' GROUP BY continent ORDER BY total_deaths_per_million DESC LIMIT 6;", connexion)

data = list(df['total_deaths_per_million'])
plt.bar(list(df['continent']), data)
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
